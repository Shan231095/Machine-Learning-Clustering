```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
```

```python
X=dataset.iloc[:,[3,4]]. values 
```

```python
from sklearn.cluster import SpectralClustering
clustering = SpectralClustering(n_clusters=2, assign_labels='discretize', random_state=0)
y_SC=clustering.fit_predict(X)
```

```python
supervised=pd.DataFrame(dataset)
```

```python
supervised['Cluster_group']=y_SC
```

```python
import seaborn as sns
```

```python
facet=sns.lmplot(data=supervised, x=supervised.columns[3],
                 y=supervised.columns[4], 
                 hue=supervised.columns[5], 
                 fit_reg=False,
                 legend = True,
                 legend_out=True)
```
